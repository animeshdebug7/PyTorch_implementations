```python
import torch.nn as nn
L = nn.CrossEntropyLoss()
# y_true = torch.tensor([[1.,0.]])
y_true = torch.tensor([1])

x = torch.tensor([1.,2.])
w1 = torch.randn([2,3], requires_grad = True)
b1 = torch.ones([1,3], requires_grad = True)
h = torch.sigmoid((torch.matmul(x, w1)) + b1)
w2 = torch.randn([3,2], requires_grad = True)
b2 = torch.ones([1,2], requires_grad = True)
y = (torch.matmul(h, w2)) + b2
loss = L(y, y_true)
```

```python
y
```

```python
y_true.shape
```

```python
y.shape
```

```python
loss.backward()
w1.grad
```

```python
w2.grad
```

```python
b1.grad
```

```python
b2.grad
```

```python
#verify
print(torch.softmax(y, dim=1))
```

```python
#clear gradients to handle the accumulation
w1.grad.zero_()
w2.grad.zero_()
b1.grad.zero_()
b2.grad.zero_()
```
