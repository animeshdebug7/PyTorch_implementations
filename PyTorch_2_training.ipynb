```python
# NN
```

```python
class NN():
    def __init__(self, X):
        self.weights = torch.rand(X.shape[1], 1, dtype = torch.float64, requires_grad = True)
        self.bias = torch.zeros(1, dtype = torch.float64, requires_grad = True)

    def fwd(self, X):
        z = torch.matmul(X, self.weights) + self.bias
        y = torch.sigmoid(z)
        return y
```

```python
lr = 0.1
epochs = 500
```

```python
model = NN(X_train)
model.weights
```

```python
model.bias
```

```python
for epoch in range(epochs):
    # fwd
    y = model.fwd(X_train)
    y_train = y_train.to(torch.float64)
    y = y.flatten()
    # loss
    bce = nn.BCELoss()
    loss = bce(y, y_train)
    # back
    loss.backward()
    # grad = model.weights.grad
    # update
    with torch.no_grad():
        model.weights -= lr * model.weights.grad
        model.bias -= lr * model.bias.grad

    model.weights.grad.zero_()
    model.bias.grad.zero_()

    print(f'epoch: {epoch + 1}, loss: {loss.item()}')
```

```python
with torch.no_grad():
    y = model.fwd(X_test)
    y = (y > 0.5).float()

accuracy = (y == y_test).float().mean()
accuracy
```

```python
class Datass(Dataset):
    def __init__(self, X, y):
        self.X = torch.tensor(X, dtype = torch.float32)
        self.y = torch.tensor(y, dtype = torch.long)

    def __len__(self):
        return len(self.X)

    def __getitem__(self, idx):
        return self.X[idx], self.y[idx]
```

```python
class ANN(nn.Module):
    def __init__(self, n):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(n, 128),
            nn.BatchNorm1d(128),
            nn.ReLU(),
            nn.Linear(128, 64),
            nn.BatchNorm1d(64),
            nn.ReLU(),
            nn.Linear(64, 10)
        )

    def forward(self, X):
        return self.net(X)
```

# Hyperparameter tuning (Optuna)

```python
class NNH(nn.Module):
    def __init__(self, input_dim, output_dim, num_hidden, neurons):
        super().__init__()

        layers = []
        for i in range(num_hidden):
            layers.append(nn.Linear(input_dim, neurons))
            layers.append(nn.BatchNorm1d(neurons))
            layers.append(nn.ReLU())
            layers.append(nn.Dropout(0.2))
            input_dim = neurons
        layers.append(nn.Linear(neurons, output_dim))

        self.model = nn.Sequential(*layers)

    def forward(self, X):
        return self.model(X)
```

```python
def obj(trial):
    
    # next hyperparams
    num_hidden = trial.suggest_int("num_hidden", 1, 5)
    neurons = trial.suggest_int("num_neurons", 8, 128, step = 8)
    # model init
    input_dim = 784
    output_dim = 10

    model = NNH(input_dim, output_dim, num_hidden, neurons)
    
    # params init
    lr = 0.1
    epochs = 100
    # optim selection
    crit = nn.CrossEntropyLoss()
    optimizer = torch.optim.SGD(model.parameters())
    # train loop
    for epoch in range(epochs):
        for batch_X, batch_y in train_load:
            #data to gpu
            # batch_X = batch_X.to(device)
            # batch_y = batch_y.to(device)
    
            #fwd
            y = net(batch_X)
            
            #loss
            loss = crit(y, batch_y)
            #back
            optimizer.zero_grad()
            loss.backward()
            #update param
            optimizer.step()

    # eval
    model.eval()
    total = 0
    correct = 0
    with torch.no_grad():
        for batch_X, batch_y in test_load:
            outputs = model(batch_X)
            _, predicted = torch.max(outputs, 1)
            total = total + batch_y.shape[0]
            correct = correct + (predicted == batch_y).sum().item()
            accuracy = correct / total
    
    return accuracy
```

```python
import optuna

study = optuna.create_study(direction = "maximize")
```

```python
study.optimize(obj, n_trials = 10)
```
